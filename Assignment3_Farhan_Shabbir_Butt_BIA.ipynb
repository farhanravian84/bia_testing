 **Question 1**  
 Write a function to calculate the mode of a list of numbers without using any external libraries. 
The function should handle cases with multiple modes by returning all modes as a list.

```python
list_mode=[1,1,1,1,2,2,2,2,3]
def mode_out(list):
    iter_dict={}
    
    for x in list:
        if x in iter_dict:
            iter_dict[x]+=1
        else:
            iter_dict[x]=1
    max_count=0
    mode=[]
    for value in iter_dict.values():
        if value>max_count:
            max_count=value
    
    for key in iter_dict:
        if iter_dict[key]==max_count:
            mode.append(key)
    return mode
mode_out(list_mode)
```

 Write a function that takes a list of integers and returns a new list sorted in ascending order, 
without using any built-in sorting functions. 
Can’t Use .sort function

```python
list_num=[9,8,7,6,5,4,3,2,0,1]
def num_sort(list):
    for i in range(len(list)):
        for j in range(i+1,len(list)):
            if list[i]>list[j]:
                list[i],list[j]=list[j],list[i]
        
    return list
num_sort(list_num)
```

**Question7**  
 Write a function that checks if two strings are anagrams of each other.No use of built-in function

```python
str1="listen"
str2="silent"
def anagram_check(str1,str2):
    if len(str1)!=len(str2):
        return "Not anagrams."
    else:
        
        str1_freq={}
        str2_freq={}
        
        for x in str1:
            if x in str1_freq:
                str1_freq[x]+=1
            else:
                str1_freq[x]=1
        for y in str2:
            if y in str2_freq:
                str2_freq[y]+=1
            else:
                str2_freq[y]=1
        for z in str1_freq:
            if z not in str2_freq:
                return "Not anagrams."
            elif str1_freq[z]!=str2_freq[z]:
                return "Not anagrams."
        return "The strings are anagrams."
            
anagram_check(str1,str2)
    
```

**Question8**  
 Write a function that finds the missing number in a list of integers from 1 to n.

```python
list_int=[1,2,4,5,6,9]
def missing_num(list):
    missing_numbers=[]
    max_number=1
    for x in list:
        if x>max_number:
            max_number=x
    for i in range(1,max_number+1):
        if i in list:
            continue
        else:
            missing_numbers.append(i)
    if len(missing_numbers)==0:
        return "No missing number found."
    else:
        print("Missing numbers are:")
        return missing_numbers
missing_num(list_int)   
    
```

**Question9**  
 Write a function that takes a string of text and returns the n most frequent words in the text.

```python
str_text="The quick brown fox, jumps over the lazy dog. The dog barks!"
def freq_words(string,n):
    new_str=string.lower()
    new_str=new_str.replace(",","").replace(".","").replace("!","").replace("?","")
    split_str=new_str.split()
    str_freq={}
    for word in split_str:
        if word in str_freq:
            str_freq[word]+=1
        else:
            str_freq[word]=1
    reqd_word=[]
    
    str_freq=sorted(str_freq,key=lambda x:str_freq[x],reverse=True)
    
    for i in range(n):
        reqd_word.append(str_freq[i])
    return reqd_word
    
freq_words(str_text,2)
```
